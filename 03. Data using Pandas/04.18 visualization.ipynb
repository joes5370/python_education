```python
##차트 이름과 x축 y축 이름넣기
x = list(range(1,11,1))
y=[]
for i in x:
    y.append(i**2)
plt.plot(x,y,"go--")
plt.xlabel('index')
plt.ylabel('value')
plt.title('simple plot')
```

```python
#특정 구간 확대
x = list(range(1,11,1))
y=[]
for i in x:
    y.append(i**2)
plt.plot(x,y,"ro--")
plt.xlim(3,7)
plt.ylim(10,40)
```

```python
%matplotlib inline
```

```python
#레전드 설정
x = list(range(1,11,1))
y = [i**2 for i in x]
#한줄에 for문을 돌려 y의 list에 넣기

y2 = [i**2.2 for i in x]
plt.plot(x,y,"go--",label = "first")
plt.plot(x,y2, "bo--", label = "second")
plt.legend(loc=0)
```

```python
%matplotlib inline
```

```python
#어노테이션
x = list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]
plt.plot(x,y,"go--",label = "first")
plt.plot(x,y2, "bo--", label = "second")
plt.legend(loc=0)
plt.annotate('point',xy=(6,40),
            xytext=(4,4),arrowprops = dict(color = "blue",width = 2))
```

### 데이터 시각화 고급 

### 1. 서브플롯 그리기 

```python
x = list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
#1행에 2열을 만들 건데 그중 첫번째
plt.plot(x,y, 'b', lw= 1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc = 0)
plt.ylabel('value')
plt.title('supplt')

ax2 = plt.subplot(122)
#1행에 2열을 만들 건데 그중 두번째
plt.plot(x,y2, 'g', lw = 1.5, label = '2nd')
plt.plot(x,y2, ' go')
plt.grid(True)
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
```

### 이중축 활용하기

```python
x = list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'b--', lw= 1.5, label = '1st')
#lw는 선굵기
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc = 0)
plt.ylabel('value')
plt.title('supplt')

ax2 = ax1.twinx()
plt.plot(x,y2, 'g', lw = 1.5, label = '2nd')
plt.plot(x,y2, ' go')
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
x = (Data2015["WEEK"])
y = (Data2015["QTY"])
y2 = (Data2015["PRO_PERCENT"])

fig = plt.figure(figsize = (10,5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'bo--', lw = 1.5)
plt.grid(True)
plt.ylabel("QTY")
plt.title("supplot")

ax2 = ax1.twinx()
plt.plot(x,y2, 'g',lw = 1.5, label = '2nd')
plt.legend(loc = 0)
plt.ylabel("pro_percent")
```
