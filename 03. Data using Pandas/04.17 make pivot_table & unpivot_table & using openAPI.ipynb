### pivot테이블 만들기 행을 열로

```python
selloutData = pd.read_csv('../Python_FinTech/dataset/kopo_channel_seasonality_new.csv')
```

```python
pivotResult = selloutData.\
pivot_table(index = ['REGIONID','PRODUCT'],
           columns = 'YEARWEEK',values = 'QTY',
           aggfunc = "mean")
pivotResult2 = pivotResult.reset_index()
pivotResult2.head()
```

```python
pivotResult2.to_csv('./pivotResult.csv')
```

### unpivot만들기(피벗한 테이블을 다시 원상 복귀 시키기) 열을 행으로  = 컴퓨터와 대화할 수 있는 형태
#### (외부에 데이터를 공시할 때는 unpivot_table로 올려야함)

```python
unpivotResult = pd.melt(pivotResult2, id_vars=['REGIONID','PRODUCT'],var_name = 'YEARWEEK',value_name='QTY')
unpivotResult.head()
```

```python
unpivotResult.to_csv('./unpivotResult.csv')
```
