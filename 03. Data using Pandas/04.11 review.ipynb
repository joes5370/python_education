


### 53주차 제거 복습 

```python
import pandas as pd
import numpy as np
selloutData = pd.read_csv('../dataset/kopo_channel_seasonality_new.csv')
selloutData.head()
```

```python
selloutData["NEW_QTY"] = np.where(selloutData["QTY"]<0,0,selloutData["QTY"])
```

```python
selloutData["WEEK"] = (selloutData["YEARWEEK"]%100).astype(int)
```

```python
selloutData["YEAR"] = round(selloutData["YEARWEEK"]/100).astype(int)
selloutData.head()
```

```python
refineData = selloutData[selloutData.WEEK < 53]
refineData.head()
```

### 함수로 만들어 버리기 

```python
def customF(inputColume):
    yearValue = inputColume/100
    yearValue = int(yearValue)
    return yearValue
```

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].apply(customF)
selloutData.head()
```

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
sortData = refineData.sort_values(sortKey,ascending = True)
sortData.head()
```

###  인덱스 정리

```python
step1Data = sortData.reset_index(drop = True)
step1Data.head()
```

###  rolling 함수 다시 써보기

```python
## 롤링함수를 쓰기전에는 정렬은 기본이다!!
## 정렬은 모든 단계의 기본적으로 하는것
```

### 기본 이동평균 함수

```python
step1Data["MA"] = step1Data.NEW_QTY.rolling(window = 5,min_periods = 1 ,center = True).mean()
step1Data.head()
## min_periods는 앞에서 부터 나오게 안쓴다면 window값에 따라 MA가 빠져서 나온다.
```

```python
step1Data.to_csv('./reviewResult.csv')
# 현재 위치에 정리한 파일 보내기
```

### groupby사용 그룹지정 

```python
groupKey = ["REGIONID","PRODUCT","YEAR"]
```

```python
groupResult = step1Data.groupby(groupKey)[["NEW_QTY"]].mean()
## 그룹키를 묶어 그룹키에 대한 NEW_QTY의 평균을 나타내라
```

```python
groupResult.head()
```

```python
g2 = groupResult.reset_index()
g2.head()
```

```python
result = pd.read_csv('./reviewResult.csv')
result["YEAR"] = (result.YEARWEEK/100).astype(int)
```

```python
r2 = result.reset_index()
r2.head()
```

###  데이터 merge

```python
mergeData = pd.merge(r2,g2,on = groupKey, how = "left")
# how의 left,right,inne는  병합할 데이터 두개 기준
mergeData.head()
```

```python
mergeData2 = pd.merge(r2,g2,how = "left", on = ["REGIONID","PRODUCT"])
mergeData2.head()
```

###  그룹바이 함수 만들기

```python
groupKey2 = ["REGIONID","PRODUCT"]
```

```python
groupDataSet = mergeData.groupby(groupKey2)
```

```python
oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[0])
oneGroup.head()
```

```python
def groupRolling(oneGroup):
    indexGroupData = oneGroup.reset_index(drop = True)
    indexGroupData["MA2"] = indexGroupData["NEW_QTY_x"].rolling(window = 5, center = True).mean()
    return indexGroupData
```

```python
finalResult = mergeData.groupby(groupKey2).apply(groupRolling)
```
