### 7.데이터 조작하기

```python
csData = pd.read_csv('../dataset/customerdata.csv')
csData.head()
```

```python
import numpy as np
```

```python
csData["PRODUCTAGE_AGE_NEW"] = np.where(csData["PRODUCTAGE"]<1,1,\
                                        np.where(csData["PRODUCTAGE"]<2,2,\
                                                 np.where(csData["PRODUCTAGE"]<3,3,\
                                                          5)))
csData.head()
```

### Quiz1. 불량데이터 처리

```python
selloutData["QTY_NEW"] = np.where(selloutData["QTY"]<0,0,\
                                 selloutData["QTY"])
selloutData.head()
```

### Quiz2. 데이터 통합(53주차제거)

```python
selloutData["YEAR"] = selloutData.YEARWEEK.astype(str)
selloutData["WEEK"] = selloutData.YEARWEEK.astype(str)
selloutData["YEAR"] = selloutData['YEARWEEK'].str[0:4]
selloutData["WEEK"] = selloutData['YEARWEEK'].str[4:]
```

```python
refineSelloutData = selloutData[selloutData.WEEK.astype(int) <53]
refineSelloutData.head()
```

### 8. 이상데이터 정제하기

```python
def refining(df):
    if df['PRODUCTAGE'] <1 :
        return 1
    elif df['PRODUCTAGE'] <2:
        return 2
    else:
        return 5
csData["PRODUCTAGE_NEW"] = csData.apply(refining,axis =1)
csData
```

### 10.데이터요약 (groupby)

```python
meanValue = csData.\
groupby(['CUSTTYPE2','EMI'],as_index = True).\
mean()
meanValue
meanResult = meanValue.reset_index()
meanResult.head()
```

### 11.데이터요약 심화과정(groupby) 

```python
groupData = sortedSelloutData.\
groupby(["REGIONID","PRODUCT"]).agg(['mean','std'])['QTY_NEW']
#그룹을 REGIONID,PRODUCT묶고 QTY_NEW의 평균과 표준편차를 보여주세요
groupData.head()
```

```python
groupResult = groupData.reset_index()
#그룹으로 묶어준 후 끝나면 항상 최초 데이터로 만들어주기
groupResult.head()
```
