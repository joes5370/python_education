### 4.데이터 조작하기

```python
csDATA2 = pd.read_csv("../dataset/customerdata.csv")
```

```python
import numpy as np
```

```python
csDATA2["PRODUCTAGE_age_NEW"] = np.where(csDATA2["PRODUCTAGE"]<1,1,\
                            np.where(csDATA2["PRODUCTAGE"]<2,2,\
                            np.where(csDATA2["PRODUCTAGE"]<3,3,\
                                    5)))
csDATA2.head(5)
```

### 4.이상데이터 정제하기 

```python
# 컬럼 추가 생성 및 조건에 따른 컬럼명 변경
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] <2:
        return 2
    else:
        return 5
csDATA2["PRODUCTAGE_NEW2"] = csDATA2.apply(refining,axis=1)
csDATA2
```
